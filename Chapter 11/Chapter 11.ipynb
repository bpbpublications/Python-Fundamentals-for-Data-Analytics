### Bar plot

```python
# x-axis
x = ['United States', 'Chine', 'Japan', 'Germany', 'India']
# y-axis
y = [23315020.56, 17734062.65, 4940877.78, 4259934.91, 3176295.07]
# Function to plot
bars = plt.bar(x, y)
# Add title and axis names
plt.title('Top 5 countries by GDP in 2021')
plt.xlabel('Countries')
plt.ylabel('GDP in Million Dollars')
# Add color for each bar
color = ['r', 'g', 'b', 'y', 'm']
for i in range(len(bars)):
    bars[i].set_color(color[i])
# function to show the plot
plt.show()
```

### Histogram

```python
# import libraries
import matplotlib.pyplot as plt
import numpy as np

# Generate random data for the histogram
data = np.random.randn(1000)

# Plotting a histogram
plt.hist(data, bins = 30, color = 'g', edgecolor = 'k')

# Adding labels and title
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Basic Histogram')

# Display the plot
plt.show()
```

### Scatter plot

```python
# Consider 2 datasets for demonstration purpose
# first dataset
x1 = [89, 43, 36, 36, 95, 88, 66, 34, 38, 20]
y1 = [21, 46, 3, 35, 67, 95, 43, 72, 58, 10]

# second dataset
x2 = [26, 29, 28, 64, 6, 5, 8, 66, 72, 40]
y2 = [26, 34, 98, 33, 38, 63, 56, 2, 47, 15]

plt.scatter(x1, y1, c = "b", linewidth = 3, marker = "s", edgecolor = "k", s = 50)
plt.scatter(x2, y2, c = "g", linewidth = 2, marker = "^", edgecolor = "k", s = 200)

plt.xlabel("X-axis")
plt.ylabel("Y-axis")

plt.show()
```

### Multiple plots

```python
import matplotlib.pyplot as plt
from math import pi

# some example data to display
x = np.linspace(0, 10, 400)
y = np.sin(x ** 2)

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2, 2)
fig.suptitle('Demonstrating Subplots')
ax1.plot(x, y, 'tab:blue')
ax2.plot(x, y*2, 'tab:orange')
ax3.plot(x, -y, 'tab:green')
ax4.plot(x, -y*2, 'tab:red')

# Using the instruction mentioned below, you can add title to each plot
ax1.set_title('First plot')
ax2.set_title('Second plot')
```

## Exploring Seaborn

```python
# installing seaborn
!pip install seaborn
```

```python
import seaborn as sns
```

```python
print(sns.get_dataset_names())
```

```python
df = sns.load_dataset('tips')
df.head()
```

```python
df.tail()
```

### Scatter plot

```python
# scatter plot of total_bill vs tip day wise
sns.scatterplot(x='total_bill', y='tip', data=df, hue='day')
plt.show()
```

### Line plot

```python
# line plot of total_bill vs tip day wise
sns.lineplot(x='total_bill', y='tip', data=df, hue='day')
plt.show()
```

### Bar plot

```python
# bar plot of day vs tip
sns.barplot(x='day', y='tip', data=df, color='#00FFF0')
```

### Box plot

```python
# box plot of day vs total_bill
sns.boxplot(x='day', y='total_bill', data=df)
plt.show()
```

### Categorical plot

```python
# categorical plot of smoker vs tip
sns.catplot(x='smoker', y='tip', data=df, color='navy')
```

### Count plot

```python
# count plot of sex
sns.countplot(x='day', data=df, hue='sex')
plt.show()
```

```python
# count plot of sex, day wise
sns.countplot(x='sex', data=df, hue='day')
plt.show()
```

### Lmplot plot

```python
# LM plot of total_bill vs tip, sex wise
sns.lmplot(x='total_bill', y='tip', data=df, hue='sex', fit_reg=False, markers=['^', 's'],
          palette='ocean', row='sex', col='smoker', height=4, aspect=1.5)
plt.show()
```

### Pair plot

```python
# pair plot based on sex
sns.pairplot(data=df, hue='sex')
plt.show()
```
