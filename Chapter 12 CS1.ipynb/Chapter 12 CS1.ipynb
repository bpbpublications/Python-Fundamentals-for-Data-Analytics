```python
# import necessary libraries
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

# dataset is availeble at https://www.kaggle.com/datasets/tawfikelmetwally/employee-dataset

```python
# read data from the csv file
df=pd.read_csv("Employee.csv")
df
```

```python
# display details of the dataset
df.info()
```

```python
# describe statistical details of the dataset
df.describe()
```

```python
# check for missing values
df.isnull().sum()
```

```python
# visualize number of male and females 
gender_counts = df['Gender'].value_counts()
plt.figure(figsize=(4, 4))

bars = plt.bar(gender_counts.index, gender_counts.values, color=['skyblue', 'salmon'])
bars[0].set_color('blue')  # Male
bars[1].set_color('yellow')   # Female

plt.title('Count of Male and Female')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.show()
```

```python
# visualize age distribution
sns.displot(df['Age'])
```

```python
# pi chart of city wise employee percentage
df['City'].value_counts().plot(kind='pie',autopct='%.2f',colormap='viridis')
```

```python
# pi chart of education qualification of employee
df['Education'].value_counts().plot(kind='pie',autopct='%.1f',colormap='plasma')
```

```python
# How long the employee is working city wise

plt.figure(figsize=(14, 8))
sns.boxplot(x='City', y='JoiningYear', data=df, hue='City')
plt.title('Variation of Length of Service Across Different Cities')
plt.xlabel('City')
plt.ylabel('Joining Year')
plt.xticks(rotation=45)
plt.show()
```

```python
# the age and gender distribution of employees

plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='Age', hue="Gender")

plt.title('Age and Gender distribution within the workforce')
plt.xlabel('Age')
plt.ylabel('Count')
plt.show()
```

```python
# understand the leave taking behaviour

leave_distribution = df['LeaveOrNot'].value_counts()

# Plotting a pie chart
plt.figure(figsize=(4, 4))
plt.pie(leave_distribution, labels=leave_distribution.index, autopct='%1.1f%%', 
        startangle=90, colors=['green', 'yellow'])
plt.title('Leave-Taking Behavior Among Employees')
plt.show()
```

```python
# leave pattern based on gender
sns.barplot(y=df['LeaveOrNot'],x=df['Gender'], hue=df['Gender'])
```

```python
# leave pattern based on city and education
sns.barplot(y=df['LeaveOrNot'],x=df['City'],hue=df['Education'])
```

```python
# leave pattern based on city and gender
sns.barplot(y=df['LeaveOrNot'],x=df['City'],hue=df['Gender'])
```
